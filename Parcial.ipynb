Â¿Como podemos agrupar los estados?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/content/sample_data/crime_data.csv')
```

```python
dataset
```

```python
State = dataset.iloc[:, [0]].values
Homicidios = dataset.iloc[:, [1]].values
Asaltos = dataset.iloc[:, [2]].values
Poblacion_area_urbana = dataset.iloc[:, [3]].values
Violaciones = dataset.iloc[:, [4]].values
```

```python
Violaciones
```

```python
X = dataset.iloc[:,[1,2,3,4]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
dataset['Cluster'] = y_kmeans
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.countplot(x='Cluster', data=dataset, palette='viridis')
plt.xlabel('Cluster')
plt.ylabel('NÃºmero de Estados')
plt.title('NÃºmero de Estados en Cada Cluster')
plt.show()
```

Se logra ver que el resultado del ejercicio son de 3 clusteres, en donde se clasifican con complejidad:

* Cluster 0: Complejidad Baja.
* Cluster 1: Complejidad Media.
* Cluster 2: Complejidad Alta.
